```python
import chardet
import langdetect
from langdetect import detect
from deep_translator import GoogleTranslator
import nltk
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
```

```python
# Function to detect encoding
def detect_encoding(text):
    detected = chardet.detect(text.encode())  # Detect encoding
    return detected['encoding']

# Function to detect language
def detect_language(text):
    try:
        return detect(text)
    except langdetect.lang_detect_exception.LangDetectException:
        return "Unknown"

# Function to translate non-English text to English
def translate_to_english(text, source_lang):
    if source_lang == "en":  # If already English, return original text
        return text

    try:
        translated_text = GoogleTranslator(source=source_lang, target="en").translate(text)
        return translated_text
    except Exception as e:
        return f"Translation Error: {str(e)}"

# Function to tokenize text
def tokenize_text(text):
    return word_tokenize(text)

# Processing Yahoo News examples
print("\n=== Handling Multiple Languages & Encoding in Yahoo Finance News ===")

for i, news in enumerate(yahoo_finance_news):
    encoding = detect_encoding(news)
    language = detect_language(news)
    print(f"\nNews {i + 1}:")
    print(f"Encoding: {encoding}")
    print(f"Language: {language}")

    if language != "en":
        translated_news = translate_to_english(news, language)
        print(f"Translated News: {translated_news}")

```
