```python
def extract_semantic_relationships(text):
  doc = nlp(text) # get the context
  relationships = []

  # Process each sentence in the text.
  for sent in doc.sents:
    subject = None
    verb = None
    obj = None
    # Iterate through tokens in the sentence
    for token in sent:
      # Get subjects
      if token.dep_ in ("nsubj", "nsubjpass") and subject is None:
        subject = token.text
      # Get the  verb.
      if token.pos_ == "VERB" and verb is None:
        verb = token.lemma_
      # Get objects
      if token.dep_ in ("dobj", "pobj") and obj is None:
        obj = token.text
      # If a complete relationship is found, save as a relationship
      if subject and verb and obj:
        relationships.append((subject, verb, obj))
  return relationships
```

```python
import spacy

# Load the spaCy English model.
nlp = spacy.load("en_core_web_sm")

test_text = "Apple acquired Beats. Microsoft released Windows. Cat mat sat"

expected_relationships = [("Apple", "acquire", "Beats"), ("Microsoft", "release", "Windows")]

# Extract relationships using our function.
result = extract_semantic_relationships(test_text)

# Print the result
print("Result:", [i for i in result])
```
