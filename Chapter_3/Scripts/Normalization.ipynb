```python
import re
import nltk
import unicodedata
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from contractions import fix  # Expands contractions like "won't" → "will not"

# Download required NLTK resources (only needed once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('stopwords')


# Example Yahoo Finance news headlines
yahoo_finance_news = [
    "Stock market surges as tech giants report strong earnings!",
    "The Fed’s interest rate hike isn't surprising, say analysts.",
    "Tesla shares dropped by 5%—but some investors remain hopeful!",
    "Cryptocurrency markets are volatile: Bitcoin falls below $40,000.",
    "Apple reports record-high quarterly revenue—growth continues!",
    "Market sentiment fluctuates; traders prepare for volatility ahead.",
]
```

```python
# Initialize tools
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
# Normalization function
def normalize_text(text):

    text = text.lower()  # Convert to lowercase
    text = fix(text)  # Expand contractions (e.g., "isn't" → "is not")
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove punctuation & special characters
    words = word_tokenize(text)  # Tokenization
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]  # Lemmatization & stopword removal
    return ' '.join(words)  # Convert list back to text
```

```python
# Apply normalization on Yahoo Finance News examples
print("\n=== Normalized Yahoo Finance News ===")
for i, news in enumerate(yahoo_finance_news):
    normalized_text = normalize_text(news)
    print(f"\nOriginal News {i+1}: {news}")
    print(f"Normalized Text {i+1}: {normalized_text}")
```
