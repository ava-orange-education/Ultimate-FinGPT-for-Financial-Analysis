```python
import nltk
import spacy
from nltk.tokenize import word_tokenize, sent_tokenize

# Download necessary NLTK data (only required once)

nltk.download('punkt_tab')

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Example Yahoo Finance news headlines
yahoo_finance_news = [
    "Stock Market Surges as Tech Giants Report Strong Earnings!",
    "Federal Reserve Announces Interest Rate Hike Amid Inflation Concerns.",
    "Tesla Shares Drop 7% After Lower-Than-Expected Q4 Earnings Report.",
    "Bitcoin Rebounds 5% Following Regulatory Approval in Major Markets.",
    "Oil Prices Decline as Global Demand Forecast Weakens.",
    "Investors Cautious Ahead of Key Economic Data Release Next Week."
]

# Function for word-level tokenization using NLTK
def nltk_word_tokenization(text):
    return word_tokenize(text)

# Function for sentence-level tokenization using NLTK
def nltk_sentence_tokenization(text):
    return sent_tokenize(text)

# Function for tokenization using spaCy
def spacy_tokenization(text):
    doc = nlp(text)
    return [token.text for token in doc]  # Word tokens

# Running tokenization on Yahoo Finance examples
print("\n=== Tokenization Results ===")

for i, news in enumerate(yahoo_finance_news):
    print(f"\nNews {i+1}: {news}")

    # Word Tokenization (NLTK)
    print(" NLTK Word Tokens:", nltk_word_tokenization(news))

    # Sentence Tokenization (NLTK)
    print("  Sentence Tokens:", nltk_sentence_tokenization(news))

    # spaCy Tokenization
    print(" spaCy Tokens:", spacy_tokenization(news))
```
